```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D

# importacion de datos
df = pd.read_csv('cars.csv')

#filtrar por marca
#df[df.manufacturer_name == 'BMW'] 
fig = plt.figure() # crear figura
#filtrar por marca, color y tipo de transmision
bmw = df[(df.manufacturer_name == 'BMW') & (df.year_produced <= 2000 )  & (df.color == 'violet')]
plt.xlabel('Price USD', fontsize=16)# set xlabel
plt.ylabel('Count', fontsize=18)# set xlabel

fig.suptitle('graph of prices and production per year', fontsize=18)# titulo de la figura
#modelacion de los datos en seaborn
sns.histplot(bmw, x ="price_usd", hue="year_produced")
```
